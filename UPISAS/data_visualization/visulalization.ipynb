```python
import pandas as pd
import numpy as np

# # Load the data from the provided CSV file
file_path="/content/baseline_snapshot_data.csv"
# file_path="/content/hybrid_snapshot_data.csv"

data = pd.read_csv(file_path)
# print(data.columns)

# Filter entries for 'ordering-service'
ordering_service_data = data[data['Service ID'] == 'ORDERING-SERVICE']

# Save the filtered data to a new CSV file
output_file_path = '/content/baseline_data.csv'
# output_file_path = '/content/hybrid_data.csv'

ordering_service_data.to_csv(output_file_path, index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the baseline and hybrid strategy data files
baseline_file_path = '/content/baseline_data.csv'
hybrid_file_path = '/content/hybrid_data.csv'

baseline_data = pd.read_csv(baseline_file_path)
hybrid_data = pd.read_csv(hybrid_file_path)

# Assume that the indices of the rows represent the data points (normalized automatically)
baseline_data['Data Point'] = range(len(baseline_data))
hybrid_data['Data Point'] = range(len(hybrid_data))

# Plot 1: Availability Comparison
plt.figure(figsize=(8, 4))
plt.plot(baseline_data['Data Point'], baseline_data['Average Availability (%)'], label='Baseline - Availability', color='blue')
plt.plot(hybrid_data['Data Point'], hybrid_data['Average Availability (%)'], label='Hybrid - Availability', color='green')
plt.title('Availability Comparison between Baseline and Our Hybrid strategy')
plt.xlabel('Iterations')
plt.ylabel('Average Availability (%)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot 2: Response Time Comparison
plt.figure(figsize=(8, 4))
plt.plot(baseline_data['Data Point'], baseline_data['Average Response Time (ms)'], label='Baseline - Response Time', color='red')
plt.plot(hybrid_data['Data Point'], hybrid_data['Average Response Time (ms)'], label='Hybrid - Response Time', color='orange')

# Add a horizontal line for the threshold (1000 ms)
plt.axhline(y=1000, color='blue', linestyle='--', label='Threshold (1000 ms)')

plt.title('Response Time Comparison between Baseline and Our Hybrid strategy')
plt.xlabel('Iterations')
plt.ylabel('Average Response Time (ms)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

```
